```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

<h2><center>Funçōes</center></h2>

```python
       
def entre_periodos(dados,inicio,fim):
    selecao = (dados.Data >= inicio) & (dados.Data <= fim)
    dados_entre_periodos = dados[selecao]
    return dados_entre_periodos
```

<h2><center>Leitura do CSV </h2></center>

```python
dados = pd.read_csv('../bot/data/WEGE3_D1.csv')
mapa = {
    dados.columns[0] : "Data",
    dados.columns[1]:"Horario",
    dados.columns[2] : "Open",
    dados.columns[3]: "High",
    dados.columns[4]:"Low",
    dados.columns[5]:"Close",
    dados.columns[6]:"Tick_Vol"
}
dados = dados.rename(columns = mapa)
dados.head()
```

```python
#Verificacao de erro
plt.figure(figsize=(40,10))
sns.heatmap(dados.isnull(),  cbar=False, cmap = 'summer')
```

```python
inicio = '2015.01.01'
fim = '2020.11.01'
dados_entre_periodos = entre_periodos(dados,inicio,fim)
dados_entre_periodos.head()
```

```python
dados_entre_periodos.to_csv('../dados/WEGE3_D1.csv', index = False)
```
