```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
class modelagem(object):
    def volatilidade(output):
        stock_return = output.pct_change()
        vol=100*stock_return.rolling(12).std() * np.sqrt(12)
        return vol
        pass
    def retorno(output):
        r1=np.log(output)- np.log(output.shift(1))
        r1 = r1*100
        return r1
        pass
    def risco(output):
        risco = output.std()
        return risco
        pass
    def index_shape(retorno,risco):
        sharpe = (retorno/risco)
        return sharpe
        pass
def ajuste(output):
    returno_maglu = modelagem.retorno(output['Close'])
    risco_maglu = modelagem.risco(output['Close'])
    vol_maglu = modelagem.volatilidade(output['Close'])
    dic = {'Data':output.Data,'Preco':output.Close,'Volatilidade':vol_maglu,'Retorno':returno_maglu,'Risco':risco_maglu}
    analisado_mglu = pd.DataFrame(data=dic)
    analisado_mglu.dropna(inplace=True)
    return analisado_mglu
```

```python
mglu = pd.read_csv('../dados/MGLU3_D1.csv')
wege = pd.read_csv('../dados/WEGE3_D1.csv')
itub = pd.read_csv('../dados/ITUB3_D1.csv')
cvc = pd.read_csv('../dados/CVCB3_D1.csv')
ciel = pd.read_csv('../dados/CIEL3_D1.csv')
azul = pd.read_csv('../dados/AZUL4_D1.csv')
```

```python
data_inicio ='2017.05.03'
def selecao(output):
    sel = output.Data >= data_inicio
    novo = output[sel]
    return novo
```

```python
mglu = selecao(analisado_mglu)
wege = selecao(analisando_wege)
itub = selecao(analisando_itub)
cvc = selecao(analisando_cvc)
ciel = selecao(analisado_ciel)
azul = selecao(analisado_azul)
```

```python
media_risco_mglu = mglu.Risco.mean()
media_risco_wege = wege.Risco.mean()
media_risco_cvc = cvc.Risco.mean()
media_risco_itau = itub.Risco.mean()
media_risco_ciel = ciel.Risco.mean()
media_risco_azul = azul.Risco.mean()

media_retorno_mglu = mglu.Retorno.mean()
media_retorno_wege = wege.Retorno.mean()
media_retorno_cvc = cvc.Retorno.mean()
media_retorno_itau = itub.Retorno.mean()
media_retorno_ciel = ciel.Retorno.mean()
media_retorno_azul = azul.Retorno.mean()

lista_risco = [media_risco_azul,media_risco_ciel,media_risco_cvc,media_risco_itau,media_risco_mglu,media_risco_wege]
lista_retorno = [media_retorno_azul,media_retorno_ciel,media_retorno_cvc,media_retorno_itau,media_retorno_mglu,media_retorno_wege]
```

```python
sharpe = []
for i in range(len(lista_retorno)):
    sharpe.append(modelagem.index_shape(lista_retorno[i],lista_risco[i]))
dic = {'Risco':lista_risco,
       'Retorno': lista_retorno,
       'Sharpe':sharpe}
index = ['azul','ciel','cvc','itau','mglu','weg']
risco_retorno = pd.DataFrame(data =dic,index = index)
risco_retorno
```

```python
fig = plt.subplots(figsize = (20,5))

for i in range(len(risco_retorno)):

    plt.scatter(lista_risco[i],100*lista_retorno[i], label=risco_retorno.index[i])

plt.legend(loc='lower right')


plt.ylabel('Retornos [%]',fontsize=15)
plt.xlabel('Risco [%]',fontsize=15)

plt.title('Diagrama Risco Retorno',fontsize=25)
plt.show()
```
